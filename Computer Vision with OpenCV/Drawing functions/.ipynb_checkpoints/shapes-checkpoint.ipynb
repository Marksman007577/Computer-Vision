```python
# cv2 can be used to draw basic geometric shapes. The following functions are used here cv.line(), cv.circle() ,
# cv.rectangle(), cv.ellipse(), cv.putText() etc. In all the above functions, you will see some common arguments as
# given below: img : The image where you want to draw the shapes color : Color of the shape. for BGR, pass it as a
# tuple, eg: (255,0,0) for blue. For grayscale, just pass the scalar value. thickness : Thickness of the line or
# circle etc. If -1 is passed for closed figures like circles, it will fill the shape. default thickness = 1 lineType
# : Type of line, whether 8-connected, anti-aliased line etc. By default, it is 8-connected. cv.LINE_AA gives
# anti-aliased line which looks great for curves.

import cv2
import numpy as np


def draw_line():
    """To draw a line, pass in the start and end coordinate"""
    # Create a black image
    img = np.zeros((512, 512, 3), np.uint8)
    # Draw a diagonal blue line with thickness of 5 px
    return cv2.line(img, (0, 0), (511, 511), color=(255, 0, 0), thickness=5)


def draw_rectangle():
    """To draw a rectangle, you need top-left corner and bottom-right corner of rectangle. This time we will draw a
    green rectangle at the top-right corner of image."""
    img = np.zeros((512, 512, 3), np.uint8)
    # Draw a rectangle with thickness of 3 px
    return cv2.rectangle(img, (384, 0), (510, 128), (0, 255, 0), 3)


def draw_circle():
    """To draw a circle, you need its center coordinates and radius. We will draw a circle inside the rectangle drawn
    above."""
    img = np.zeros((512, 512, 3), np.uint8)
    # Draw a rectangle with thickness of 3 px
    return cv2.circle(img, (447, 63), 63, (0, 0, 255), -1)


def draw_polygon():
    """To draw a polygon, first you need coordinates of vertices. Make those points into an array of shape ROWSx1x2
    where ROWS are number of vertices and it should be of type int32. Here we draw a small polygon of with four
    vertices in yellow color."""
    img = np.zeros((512, 512, 3), np.uint8)
    # Draw a polygon
    pts = np.array([[10, 5], [20, 30], [70, 20], [50, 10]], np.int32)
    pts = pts.reshape((-1, 1, 2))
    return cv2.polylines(img, [pts], True, (0, 255, 255))


def draw_ellipse():
    """To draw the ellipse, we need to pass several arguments. One argument is the center location (x,
    y). Next argument is axes lengths (major axis length, minor axis length). angle is the angle of rotation of
    ellipse in anti-clockwise direction. startAngle and endAngle denotes the starting and ending of ellipse arc
    measured in clockwise direction from major axis. i.e. giving values 0 and 360 gives the full ellipse. For more
    details, check the documentation of cv.ellipse(). Below example draws a half ellipse at the center of the image."""
    img = np.zeros((512, 512, 3), np.uint8)
    return cv2.ellipse(img, (256,256), (100,50), 0, 0, 180, 255, -1)


def draw_all_shapes():
    # Create a black image
    img = np.zeros((512, 512, 3), np.uint8)
    line = cv2.line(img, (0, 0), (511, 511), color=(255, 0, 0), thickness=5)
    rect = cv2.rectangle(img, (384, 0), (510, 128), (0, 255, 0), 3)
    circle = cv2.circle(img,(447,63), 63, (0, 0, 255), -1)
    pts = np.array([[10, 5], [20, 30], [70, 20], [50, 10]], np.int32)
    pts = pts.reshape((-1, 1, 2))
    polygon = cv2.polylines(img, [pts], True, (0, 255, 255))
    return line, rect, circle, polygon


def add_text_to_image():
    """To put texts in images, you need specify following things.
    Text data that you want to write
    Position coordinates of where you want put it (i.e. bottom-left corner where data starts).
    Font type (Check cv.putText() docs for supported fonts)
    Font Scale (specifies the size of font)
    regular things like color, thickness, lineType etc. For better look, lineType = cv.LINE_AA is recommended"""
    # Create a black image
    img = np.zeros((512, 512, 3), np.uint8)
    # We will write OpenCV on our image in white color.
    font = cv2.FONT_HERSHEY_SIMPLEX
    return cv2.putText(img, 'OpenCV', (10, 500), font, 4, (255, 255, 255), 2, cv2.LINE_AA)


# cv2 can be used to draw basic geometric shapes. The following functions are used here cv.line(), cv.circle() ,
# cv.rectangle(), cv.ellipse(), cv.putText() etc. In all the above functions, you will see some common arguments as
# given below: img : The image where you want to draw the shapes color : Color of the shape. for BGR, pass it as a
# tuple, eg: (255,0,0) for blue. For grayscale, just pass the scalar value. thickness : Thickness of the line or
# circle etc. If -1 is passed for closed figures like circles, it will fill the shape. default thickness = 1 lineType
# : Type of line, whether 8-connected, anti-aliased line etc. By default, it is 8-connected. cv.LINE_AA gives
# anti-aliased line which looks great for curves.

import cv2
import numpy as np


def draw_line():
    """To draw a line, pass in the start and end coordinate"""
    # Create a black image
    img = np.zeros((512, 512, 3), np.uint8)
    # Draw a diagonal blue line with thickness of 5 px
    return cv2.line(img, (0, 0), (511, 511), color=(255, 0, 0), thickness=5)


def draw_rectangle():
    """To draw a rectangle, you need top-left corner and bottom-right corner of rectangle. This time we will draw a
    green rectangle at the top-right corner of image."""
    img = np.zeros((512, 512, 3), np.uint8)
    # Draw a rectangle with thickness of 3 px
    return cv2.rectangle(img, (384, 0), (510, 128), (0, 255, 0), 3)


def draw_circle():
    """To draw a circle, you need its center coordinates and radius. We will draw a circle inside the rectangle drawn
    above."""
    img = np.zeros((512, 512, 3), np.uint8)
    # Draw a rectangle with thickness of 3 px
    return cv2.circle(img, (447, 63), 63, (0, 0, 255), -1)


def draw_polygon():
    """To draw a polygon, first you need coordinates of vertices. Make those points into an array of shape ROWSx1x2
    where ROWS are number of vertices and it should be of type int32. Here we draw a small polygon of with four
    vertices in yellow color."""
    img = np.zeros((512, 512, 3), np.uint8)
    # Draw a polygon
    pts = np.array([[10, 5], [20, 30], [70, 20], [50, 10]], np.int32)
    pts = pts.reshape((-1, 1, 2))
    return cv2.polylines(img, [pts], True, (0, 255, 255))


def draw_ellipse():
    """To draw the ellipse, we need to pass several arguments. One argument is the center location (x,
    y). Next argument is axes lengths (major axis length, minor axis length). angle is the angle of rotation of
    ellipse in anti-clockwise direction. startAngle and endAngle denotes the starting and ending of ellipse arc
    measured in clockwise direction from major axis. i.e. giving values 0 and 360 gives the full ellipse. For more
    details, check the documentation of cv.ellipse(). Below example draws a half ellipse at the center of the image."""
    img = np.zeros((512, 512, 3), np.uint8)
    return cv2.ellipse(img, (256,256), (100,50), 0, 0, 180, 255, -1)


def draw_all_shapes():
    # Create a black image
    img = np.zeros((512, 512, 3), np.uint8)
    line = cv2.line(img, (0, 0), (511, 511), color=(255, 0, 0), thickness=5)
    rect = cv2.rectangle(img, (384, 0), (510, 128), (0, 255, 0), 3)
    circle = cv2.circle(img,(447,63), 63, (0, 0, 255), -1)
    pts = np.array([[10, 5], [20, 30], [70, 20], [50, 10]], np.int32)
    pts = pts.reshape((-1, 1, 2))
    polygon = cv2.polylines(img, [pts], True, (0, 255, 255))
    return line, rect, circle, polygon


def add_text_to_image():
    """To put texts in images, you need specify following things.
    Text data that you want to write
    Position coordinates of where you want put it (i.e. bottom-left corner where data starts).
    Font type (Check cv.putText() docs for supported fonts)
    Font Scale (specifies the size of font)
    regular things like color, thickness, lineType etc. For better look, lineType = cv.LINE_AA is recommended"""
    # Create a black image
    img = np.zeros((512, 512, 3), np.uint8)
    # We will write OpenCV on our image in white color.
    font = cv2.FONT_HERSHEY_SIMPLEX
    return cv2.putText(img, 'OpenCV', (10, 500), font, 4, (255, 255, 255), 2, cv2.LINE_AA)
```
